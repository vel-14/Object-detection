```python
import cv2 
from ultralytics import YOLO

img_path = '/Users/velmurugan/Desktop/s106.jpg'
outpath = img_path.split('.')[0]+'_out.jpg'

image = cv2.imread(img_path)

model = YOLO("/Users/velmurugan/Desktop/@/python_works/computer vision/train2/weights/best.pt")  

results = model(image)[0]

threshold = 0.5

for result in results.boxes.data.tolist():
    x1,y1,x2,y2, score, class_id = result

    if score>=threshold:
        if int(class_id) == 0:
            cv2.rectangle(image,(int(x1),int(y1)),(int(x2),int(y2)),(0,255,0),4)
            cv2.putText(image,results.names[int(class_id)],(int(x1),int(y1-10)),
                        fontFace = cv2.FONT_HERSHEY_PLAIN, fontScale= 3, color=(0,255,0),
                        thickness=3, lineType=cv2.LINE_AA)
        else:
            cv2.rectangle(image,(int(x1),int(y1)),(int(x2),int(y2)),(0,0,255),4)
            cv2.putText(image,results.names[int(class_id)],(int(x1),int(y1-10)),
                        fontFace = cv2.FONT_HERSHEY_PLAIN, fontScale= 3, color=(255,0,0),
                        thickness=3, lineType=cv2.LINE_AA)

cv2.imwrite(outpath,image)
            
```

```python
video_path = '/Users/velmurugan/Desktop/tom-holland-and-zendaya-rookie-in-training-for-freeform-first-hd-version-360-ytshorts.savetube.me.mp4'
cap = cv2.VideoCapture(video_path)
ret,frame = cap.read()
H,W,C = frame.shape

out_video_path = video_path.split('.')[0]+'_out.mp4'
out = cv2.VideoWriter(out_video_path,cv2.VideoWriter_fourcc(*'MP4V'),int(cap.get(cv2.CAP_PROP_FPS)),(W,H))

threshold = 0.5

while ret:
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1,y1,x2,y2, score, class_id = result

        if score>=threshold:
            if int(class_id) == 0:
                cv2.rectangle(frame,(int(x1),int(y1)),(int(x2),int(y2)),(0,255,0),4)
                cv2.putText(frame,results.names[int(class_id)],(int(x1),int(y1-10)),
                            fontFace = cv2.FONT_HERSHEY_PLAIN, fontScale= 3, color=(0,255,0),
                            thickness=3, lineType=cv2.LINE_AA)
            else:
                cv2.rectangle(frame,(int(x1),int(y1)),(int(x2),int(y2)),(0,0,255),4)
                cv2.putText(frame,results.names[int(class_id)],(int(x1),int(y1-10)),
                            fontFace = cv2.FONT_HERSHEY_PLAIN, fontScale= 3, color=(255,0,0),
                            thickness=3, lineType=cv2.LINE_AA)
    
    out.write(frame)
    ret,frame = cap.read()

cap.release()
out.release()

```
